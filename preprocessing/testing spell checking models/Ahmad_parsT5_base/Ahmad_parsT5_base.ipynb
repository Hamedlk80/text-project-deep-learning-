#cleaning

```python
def clean_symbols_and_emojis(text):

    # Define allowed punctuation: Persian and some general
    allowed_punct = "،.؟؛:!"

    # Build regex pattern:
    # Pattern to match any character NOT in:
    # - Persian Unicode range \u0600-\u06FF
    # - English letters a-zA-Z
    # - Digits 0-9 and Persian digits \u06F0-\u06F9
    # - Half-space \u200c
    # - Whitespace \s
    # - Allowed punctuation
    pattern = fr"[^a-zA-Z0-9\u06F0-\u06F9\u0600-\u06FF\u200c\s{re.escape(allowed_punct)}]"

    # Remove disallowed characters (emojis, symbols, etc.)
    cleaned = re.sub(pattern, "", text)

    return cleaned
```

#Normalizing

```python
# Initialize hazm normalizer
normalizer = Normalizer()

def normalize_farsi_sentences(sentences):
    """
    Normalize a list of Farsi sentences using hazm.
    """
    normalized = []
    for sentence in sentences:
        cleaned = normalizer.normalize(sentence)
        normalized.append(cleaned)
    return normalized
```

#applying cleaning and normalizing

```python
# Clean + Normalize pipeline
cleaned_texts = [clean_symbols_and_emojis(t) for t in the_texts]
normalized_texts = [normalizer.normalize(t) for t in cleaned_texts]

# Show original, cleaned, and normalized
for i, (orig, clean, norm) in enumerate(zip(the_texts, cleaned_texts, normalized_texts), 1):
    print(f"{i}. Original   : {orig}")
    print(f"   Cleaned    : {clean}")
    print(f"   Normalized : {norm}")
    print("-" * 50)
```

# loading and testing model

```python
import torch, pandas as pd
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

device = "cuda" if torch.cuda.is_available() else "cpu"
model_name = "Ahmad/parsT5-base"

tok = AutoTokenizer.from_pretrained(model_name)
# اگر pad_token نداشت، با eos یکی کن تا ارور نده
if tok.pad_token is None:
    tok.pad_token = tok.eos_token

dtype = torch.float16 if device=="cuda" else torch.float32
model = AutoModelForSeq2SeqLM.from_pretrained(model_name, torch_dtype=dtype).to(device)
model.config.pad_token_id = tok.pad_token_id

def build_prompt(txt: str) -> str:
    # یک قالب سادهٔ دستوری برای وادار کردن مدل به بازنویسی تمیز
    return f"اصلاح املایی و نگارشی جمله زیر را انجام بده و فقط نسخهٔ اصلاح‌شده را برگردان:\n{txt}"

def generate_batch(batch_texts, max_src_len=256, max_new_tokens=192):
    enc = tok([build_prompt(t) for t in batch_texts],
              padding=True, truncation=True, max_length=max_src_len, return_tensors="pt").to(device)
    with torch.no_grad():
        out_ids = model.generate(
            **enc,
            max_new_tokens=max_new_tokens,
            num_beams=4,
            no_repeat_ngram_size=3,
            length_penalty=1.0,
            early_stopping=True
        )
    return tok.batch_decode(out_ids, skip_special_tokens=True)

# --- Load CSV and run only on Normalized ---
df = pd.read_csv("farsi_cleaned_normalized.csv")
texts = df["Normalized"].fillna("").astype(str).tolist()

batch_size = 8
outputs = []
for i in range(0, len(texts), batch_size):
    outputs.extend(generate_batch(texts[i:i+batch_size]))

df["parsT5_on_Normalized"] = outputs
df.to_csv("farsi_parsT5_corrected_on_normalized.csv", index=False, encoding="utf-8-sig")
print("✅ Saved to farsi_parsT5_corrected_on_normalized.csv")
```
